```python
import numpy as np
from pettingzoo.mpe import simple_v2, simple_reference_v2, simple_reference_v3
import gym
import supersuit as ss
from stable_baselines3.common.env_checker import check_env
from stable_baselines3 import PPO
from torch.utils.tensorboard import SummaryWriter
```

```python
game = simple_reference_v3.parallel_env(max_cycles=80)
game = ss.pettingzoo_env_to_vec_env_v0(game)
env = ss.concat_vec_envs_v0(game, 10, num_cpus=5, base_class='stable_baselines3')

model = PPO("MlpPolicy", env, verbose=2, gamma=0.95, n_steps=10, ent_coef=0.0905168, learning_rate=0.0004, vf_coef=0.042202, max_grad_norm=0.9, gae_lambda=0.99, n_epochs=3, clip_range=0.2, batch_size=10,tensorboard_log="run/reference_v3_stablebaseline_ppo")
print("done")
```

```python
# model.load("policy_reference_v3")
model.learn(total_timesteps=25*1000, log_interval=100)
model.save("policy_reference_v3")
```

```python
from stable_baselines3.common.vec_env import VecVideoRecorder, DummyVecEnv

env_id = 'CartPole-v1'
video_folder = 'logs/videos/'
video_length = 100


game = simple_reference_v3.parallel_env(max_cycles=75)
game = ss.pettingzoo_env_to_vec_env_v0(game)

# Record the video starting at the first step
```

```python
import time

model.load("policy_reference_v3")
for i in range(15):
   obs = game.reset()
   game = VecVideoRecorder(game, video_folder,
                       record_video_trigger=lambda x: x == 0, video_length=video_length,
                       name_prefix=f"random-agent-{env_id}")
   done = False
   actions = []

   for i in range(75):
      act = model.predict(obs, deterministic=True)[0]
      actions.append(str(act))
      obs, reward, done, info = game.step(act)
      game.render()
      time.sleep(0.05)
game.close()
```
